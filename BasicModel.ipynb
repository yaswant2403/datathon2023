```python
import pandas as pd
import numpy as np 
```

```python
macro = pd.read_csv("train_macro.csv")
macro = macro.drop(columns="Unnamed: 0")
```

```python
cons = pd.read_csv("clean.csv")
cons = cons.drop(columns="Unnamed: 0")
cons
```

```python
x = cons[cons["charge_off"] == 1].groupby(["mth_code"]).size()
charge_off_rates= (cons[cons["charge_off"] == 1].groupby(["mth_code"]).size() / cons.groupby(["mth_code"]).size())
charge_off_rates = charge_off_rates.fillna(0)
charge_off_rates
macro["charge_off_count"] = macro["mth_code"].map(cons[cons["charge_off"] == 1].groupby(["mth_code"]).size())
macro["c_rate"] = macro["mth_code"].map(charge_off_rates)
macro["charge_off_count"] = macro["charge_off_count"].fillna(0)
```

```python
macro["total_accs"] = macro["charge_off_count"]/ macro["c_rate"]
macro["total_accs"] = macro["total_accs"].fillna(0)
macro["total_accs"][0] = cons.groupby(["mth_code"]).size()[201801]
macro
```

```python
test = pd.read_csv("test_macro.csv")
test = test.drop(columns=["Unnamed: 0", "index"])
```

```python
#Convert Dates to month codes for macro econmoic data
test["mth_code"] = 0
for i in range(len(test)):
    st = test.Mnemonic[i].split("/")[0]
    st2 = test.Mnemonic[i].split("/")[2]
    if len(st) < 2:
        st = "0" + st
    st2 = st2 + st
    st2 = np.int64(st2)
    test["mth_code"][i] = st2

test = test.drop(columns="Mnemonic")
```

```python
test["total_accs"] = 19866 
test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score


X_train, X_test, y_train, y_test = train_test_split(macro.drop(columns=['charge_off_count', 'mth_code', 'c_rate', 'total_accs'], axis=1), macro['c_rate'], test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

rmse = mean_squared_error(y_test, y_pred, squared=False)
print('RMSE:', rmse)
```
