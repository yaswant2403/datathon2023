```python
# calculating outliers for principal amt charged off

q1 = df_charge_off["principal_amt_chrg_off"].quantile(0.25)
q3 = df_charge_off["principal_amt_chrg_off"].quantile(0.75)
iqr = q3 - q1
mean = df_charge_off["principal_amt_chrg_off"].mean()
upper_fence = q3 + 1.5*iqr
lower_fence = q1 - 1.5*iqr
print("iqr = " + str(iqr))
print("mean = " + str(mean))
print("upper fence = " + str(upper_fence))
print("lower_fence = " + str(lower_fence))
```

```python
outliers = df_charge_off[(df_charge_off["principal_amt_chrg_off"] > upper_fence) | 
                         (df_charge_off["principal_amt_chrg_off"] < lower_fence)]

valid = df_charge_off[(df_charge_off["principal_amt_chrg_off"] <= upper_fence) & 
                         (df_charge_off["principal_amt_chrg_off"] >= lower_fence)]

```
