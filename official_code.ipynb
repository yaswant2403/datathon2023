# Model

```python
import numpy as np
import matplotlib.pyplot as plt
from sksurv.nonparametric import kaplan_meier_estimator
```

```python
#changing fico scores to numerical medians

scores = [(581,600),(721,740),(761,850),(661,680),(601,620), (701,720), (621,640), (561,580), (641,660), (741,760), (681,700),(560,300)]
medians = [(tup[0] + tup[1])/2 for tup in scores]
ranges = { }
for mid in medians:
    fico_range = scores[medians.index(mid)]
    if (fico_range[0] == 560):
        ranges['<= 560'] = mid
    elif (fico_range[0] == 761):
        ranges['761+'] = mid 
    else:
        key = str(fico_range[0]) + '-' + str(fico_range[1])
        ranges[key] = mid
ranges['Exception'] = 0
ranges['Missing'] = 0
df2 = df.replace({"bank_fico_buckets_20": ranges})
```

```python
#Eliminating the 0s 

df2 = df2[df2["bank_fico_buckets_20"] != 0]
df2["bank_fico_buckets_20"].unique()
```

```python
# Filtering dataset so it's only one year, 2019. 
# We will use this year of data to train our model.

df_2019 = df2[df2["mth_code"] > 201900]
df_2019["mth_adj"] = df2["mth_code"] - 201900
df_2019
```

```python
# Basic Kaplan-meier model

time, survival_prob1 = kaplan_meier_estimator(df_2019["charge_off"].astype(bool), df_2019["mth_adj"])
time, survival_prob1
```

```python
# Plotting Kaplan-meier model

plt.step(time, survival_prob1, where="post")
```

```python
# Kaplan-meier model, stratified on fico credit score
#Storing the survival probabilities for each credit score in data2

data2 = []
for fico in (medians):
    mask_score = df_2019["bank_fico_buckets_20"] == fico
    time, survival_prob2 = kaplan_meier_estimator(
        df_2019["charge_off"].astype(bool)[mask_score],
        df_2019["mth_adj"][mask_score])
    d = {"fico": fico, "survival_prob2": survival_prob2}
    data2.append(d)
    plt.step(time, survival_prob2, where="post",
             label="score = %s" % fico)
plt.ylabel("est. probability of no charge-off $\hat{S}(t)$")
plt.xlabel("months since start date $t$")
plt.legend(loc="best")
plt.suptitle("Kaplan-Meier Model for Bank Charge-Offs", fontweight = "bold")
plt.title("Stratified by FICO credit score range")
```

```python
# putting fico scores and their survival rates in a dataframe

fico_rates = pd.DataFrame(data2)
fico_rates = fico_rates.sort_values(by = "fico", ascending = True).reset_index()
fico_rates
```
